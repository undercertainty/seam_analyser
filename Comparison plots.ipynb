# Feedback comparator

Use this notebook to plot the graphs for two presentations against each other for comparison.

```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# Set the filename and plot title name for the first plot

seamFile1_xls='random SEaM data 1.xlsx'

plotTitle1_str='2012 Q'
```

```python
# Set the filename and plot title name for the first plot

seamFile2_xls='random SEaM data 2.xlsx'

plotTitle2_str='2014 V'
```

```python
# Set to False for absolute values rather than percentages

PLOT_PERCENT_BOOL=True
```

**Now run all the cells**

```python
feedback1_df=pd.read_excel(seamFile1_xls, sheetname=2)
feedback2_df=pd.read_excel(seamFile2_xls, sheetname=2)
```

```python
questions_ls=feedback1_df.columns
```

```python
# Plot for feedback question 13

q=0

data1_ss=feedback1_df[questions_ls[q]].value_counts()
data1_ss.drop('6. Not applicable/used', inplace=True, errors='ignore')
data1_ss.drop('Student did not answer this question', inplace=True, errors='ignore')
data1_ss.sort_index(ascending=False, inplace=True)
if PLOT_PERCENT_BOOL:
    data1_ss=data1_ss*(100/sum(data1_ss))

data2_ss=feedback2_df[questions_ls[q]].value_counts()
data2_ss.drop('6. Not applicable/used', inplace=True, errors='ignore')
data2_ss.drop('Student did not answer this question', inplace=True, errors='ignore')
data2_ss.sort_index(ascending=False, inplace=True)
if PLOT_PERCENT_BOOL:
    data2_ss=data2_ss*(100/sum(data2_ss))

upperLimit_i=max(max(data1_ss.values),
                 max(data2_ss.values))
upperLimit_i=5*(int(upperLimit_i/5)+1)

plt.suptitle(questions_ls[q])

plot1_ax=plt.subplot(121)

data1_ss.plot(kind='bar', title=plotTitle1_str, ylim=[0, upperLimit_i], ax=plot1_ax)

plot2_ax=plt.subplot(122)

data2_ss.plot(kind='bar', title=plotTitle2_str, ylim=[0, upperLimit_i], ax=plot2_ax)

plt.show()
```
